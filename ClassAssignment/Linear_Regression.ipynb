Necessary Functions

```python
def cost_function(rent):
    return (rent - 15000)**2  # Adjust this function based on your specific objective

def gradient(rent):
    return 2 * (rent - 15000)  # Adjust the gradient based on your specific objective

def gradient_descent(initial_rent, learning_rate, iterations):
    rent_values = np.zeros(iterations)
    cost_values = np.zeros(iterations)
    rent = initial_rent

    for i in range(iterations):
        cost = cost_function(rent)
        rent_values[i] = rent
        cost_values[i] = cost

        rent -= learning_rate * gradient(rent)

    return rent_values, cost_values
```

Plot

```python
initial_rent = 10000  # Set an initial value based on your dataset
learning_rate = 0.001  # Adjust the learning rate based on your specific scenario
iterations = 100

rent_values, cost_values = gradient_descent(initial_rent, learning_rate, iterations)

plt.style.use('classic')

# Plot with a colormap
plt.scatter(rent_values, cost_values, c=range(iterations), cmap='viridis', marker='o')
plt.plot(rent_values, cost_values, linestyle='-', color='black')
plt.title('Gradient Descent for Rent Optimization')
plt.xlabel('Rent values')
plt.ylabel('Cost values')
plt.colorbar(label='Iteration')
plt.grid(True)
plt.show()
```
